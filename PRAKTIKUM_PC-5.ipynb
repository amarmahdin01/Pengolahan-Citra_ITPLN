### Peng-ambangan Dwi Aras

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
daun = cv2.imread('daun.jpg', 0)
tinggi, lebar = daun.shape
```

```python
plt.imshow(daun, cmap = 'gray')
```

##### Peng-ambangan Dwi Aras dengan perkiraan nilai Ambang

```python
nilai_ambang = 128
daun_hasil = daun.copy()

for x in range(tinggi):
    for y in range(lebar):
        if daun[x, y] < nilai_ambang:
            daun_hasil[x,y] = 0
        else:
            daun_hasil[x,y] = 255
```

```python
fig, axes = plt.subplots(1, 2, figsize = (15, 11))
ax = axes.ravel()

ax[0].imshow(daun, cmap = 'gray')
ax[0].set_title('Sebelum Peng-ambangan')
ax[1].imshow(daun_hasil, cmap = 'gray')
ax[1].set_title('Setelah Peng-ambangan')
```

##### Menentukan nilai Ambang suatu gambar

```python
def titeratif(image):
    tinggi, lebar = image.shape
    t0 = 127
    while(True):
        rata_kiri = 0;
        rata_kanan = 0;
        jum_kiri = 0;
        jum_kanan = 0;
        
        for x in range(tinggi):
            for y in range(lebar):
                if(image[x, y] <= 127):
                    rata_kiri += image[x, y]
                    jum_kiri += 1
                else:
                    rata_kanan += image[x, y]
                    jum_kanan += 1
        rata_kiri /= jum_kiri
        rata_kanan /= jum_kanan
        
        t1 = (rata_kiri + rata_kanan)/2
        if((t0 - t1) < 1):
            break
            
        t0 = t1
    return round(t1)
```

```python
titeratif(daun)
```
